```python
# Requirements
import pandas as pd
```

```python
INPUT_PATH = r'D:\Kike\Proyectos\Desarrollando\electrical_data\inputs'
INPUT_FILE = '/export_historico.csv'
OUTPUT_PATH = r'D:\Kike\Proyectos\Desarrollando\electrical_data\output'
OUTPUT_FILE = '/export_historico_format.csv'
```

```python
df = pd.read_csv(INPUT_PATH + INPUT_FILE, encoding='utf-8', delimiter=';')
df = df[df['geoname']=='Península']
df['day'] = df['datetime'].str[:10]
df['day'] = pd.to_datetime(df['day'])
df['hour'] = df['datetime'].str[11:13]
df['aux_hour'] = df['hour'].astype('int') +  1
df['aux_hour'] = df['aux_hour'].astype('str')
df['aux_hour'] = df['aux_hour'].str.zfill(2)
df['hour'] = df['hour']+'-'+df['aux_hour']
df = df.rename(columns={"value": "price"})
df
```

```python
df_final = df[['hour','day','price']]
df_final['is_cheap'] =''
df_final['under_avg'] ='' # its possible calculate
df_final['market'] ='PVPC'
df_final['units'] ='€/Mwh'
del df
df_final.info()
```

```python
#Export to csv
df_final.to_csv(OUTPUT_PATH + OUTPUT_FILE, index=False, encoding='utf-8')
```
